```python
# Importation des librairies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Creation du jeu de donnees

```python
X,y=make_blobs(n_samples=1000, centers=4,n_features=2, random_state=0)
print(X.shape)
# Reshape y
y=y.reshape(-1,1)
# Plot with different colors for each class
plt.scatter(X[:,0],X[:,1],s=50, c=y, cmap='rainbow')
```

Creation du modele

```python
# Creation du modele de kmeans
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4,  random_state=0)
```

```python
model.fit(X)
```

```python
# Prediction
label = model.labels_

y_pred = model.predict(X)

# Visualisation
plt.scatter(X[:,0],X[:,1],s=50, c=label, cmap='rainbow')
```

```python
# Get the centers 
centers = model.cluster_centers_

# Plot shoiwng the centers
plt.scatter(X[:,0],X[:,1],s=50, c=label, cmap='rainbow')
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)

# Le code du prof
# plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c=np.unique(y_pred), s=100, marker='s')
```

Methode du coude (Elbow Method) pour trouver le nombre de cluster

```python
# On cree une liste de valeur de k (Nombre de cluster par le methode d'inertie pour avoir le meilleur nombre de cluster)
k_rng = range(1,10)
inertia = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(X)
    inertia.append(km.inertia_)
plt.xlabel('Nombre de clusters')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,inertia)
```
