```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
```

```python
def graphique(dataset):
    n=len(dataset)
    plt.figure(figsize=(12,8))
    for k,i in zip(dataset.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
    
```

```python
graphique(dataset)
```

```python
import scipy
```

```python
x = np.linspace(0,10,10)
y= x**2
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y, kind='linear')

new_x = np.linspace(0,10,30)
result = f(new_x)

plt.scatter(x,y)
plt.scatter(new_x, result, c='r')
```

```python
x = np.linspace(0,2,100)
y= 1/3*x**3 - 3/5*x**2 +2 + np.random.randn(x.shape[0])/20

plt.scatter(x,y)
```

```python
def f(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
from scipy import optimize
params, param_cov = optimize.curve_fit(f,x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,f(x,params[0],params[1],params[2],params[3]), c='g', lw=3)
```

```python
x = np.linspace(0,20,100)
y= x+ 4*np.sin(x) + np.random.randn(x.shape[0])

plt.plot(x,y)
```

```python
from scipy import signal

new_y = signal.detrend(y)

plt.plot(x,y)
plt.plot(x,new_y)
```

```python
x= np.linspace(0,30,1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)
plt.plot(x,y)
```

```python
from scipy import fftpack

fourier = fftpack.fft(y)

power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), power)
```
